<h1><strong>Data Visualization Using Python</strong></h1>

<h2><strong>Loading The Required Libraries</strong></h2>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

<h2><strong>Line Graph</strong></h2>

```python
#PLOTTING A LINE GRAPH USING MATPLOTLIB
month = ['jan','feb','mar','apr']
sales = [200,220,215,240]
plt.plot(month,sales,marker ='o',ls=':',lw=4)
plt.xlabel('month')
plt.ylabel('sales')
plt.title('monthly sales over a year')
plt.grid(True)
plt.show()
```

<h2><strong>Bar Chart</strong></h2>

```python
#PLOTTING A BAR CHART USING MATPLOTLIB
regions = ['north','east','west']
products_sold = [150,180,200]
plt.bar(regions,products_sold,color='orange')
plt.title('products sales by region')
plt.show()
```

```python
#PLOTTING A HISTOGRAM USING MATPLOTLIB
age = [22,25,27,30]
plt.hist(age,color='purple',edgecolor='black')
plt.title('Age distribution')
plt.grid(axis='y')
plt.show()
```

<h2><strong>Pie Chart</strong></h2>

```python
#PLOTTING A PIE CHART
brands = ['brand a','brand b','brand c']
market_share = [35,25,20]
plt.pie(market_share,labels=brands)
plt.show()
```

<h2><strong>SubPlots in Python</strong></h2>

```python
#HOW TO ADD A SUBPLOT
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)# 1row and 2 columns first plot
plt.plot(x,y)
plt.title("SALES")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2) # 1 row and 2 columns second plot
plt.plot(x,y)
plt.title("INCOME")

plt.suptitle("MY SHOP")
plt.show()
```

DATA VISUALIZATION BEST PRACTICES

1. Know Your Audience
Tailor your visuals to your audiences familiarity with the subject.
For technical audiences, go deeper into data details, while non-technical audiences benefit from simpler visualizations.
2. Choose the Right Chart Type
Comparison: Use bar charts or column charts.
Distribution: Use histograms, box plots, or violin plots.
Relationships: Use scatter plots or bubble charts.
Composition: Use pie charts (sparingly), stacked bars, or treemaps.
3. Simplify and Avoid Clutter
Remove unnecessary gridlines, borders, and decorative elements.
Use minimal colors and avoid patterns or textures that can make charts visually busy.
4. Use Colors Effectively
Use color to highlight important data points or differences.
Stick to a limited color palette and use colors with sufficient contrast.
Avoid red-green combinations for accessibility (color-blind friendly).
5. Label Clearly
Use clear titles, axis labels, and legends to help viewers understand what theyre looking at.
Avoid jargon or abbreviations unless your audience is familiar with them.
Label data points or add annotations where specific data points are important.
6. Focus on Data Integrity
Start axes at zero when possible, especially for bar charts, to avoid misleading viewers.
Use appropriate scales (e.g., log scale for exponential growth).
Dont manipulate data or visual scales in a way that distorts the datas message.
7. Tell a Story
Use your visualization to guide viewers toward insights or a clear takeaway.
Add annotations to highlight significant points or trends.
Organize visualizations in a logical order to help readers follow the narrative.
8. Consider Accessibility
Use accessible colors and large enough fonts.
Ensure your chart is readable when printed in grayscale if necessary.
Provide descriptions or labels for all elements to support screen readers if used digitally.
9. Test and Get Feedback
Share your visualizations with others to ensure they interpret the data as intended.
Make adjustments based on feedback to improve clarity and effectiveness.
